```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("svr.csv")
X = dataset.iloc[:, 1:-1].values  # .valuesデータ型をvaluesに変える
y = dataset.iloc[:, -1].values
```

```python
X
```

```python
y
```

```python
y = y.reshape(len(y), 1)
y
```

フィーチャースケーリング

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X = sc_X.fit_transform(X)
y = sc_y.fit_transform(y)
```

```python
X
```

```python
y
```

## データセットを使ったモデルの訓練 

```python
from sklearn.svm import SVR
regressor = SVR(kernel='linear', epsilon=2)
regressor.fit(X, y)
```

結果の可視化

```python
plt.scatter(sc_X.inverse_transform(X), sc_y.inverse_transform(y), color="red")
plt.plot(sc_X.inverse_transform(X).reshape(-1,1), sc_y.inverse_transform(regressor.predict(X).reshape(-1,1)), color="blue")
plt.show()
```
